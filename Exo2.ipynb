```python
from sklearn.neural_network import MLPRegressor
import matplotlib.pyplot as plt
import numpy as np
```

```python
f = lambda x: [[x] for x in x]

noise_level = 0.1
X_train = np.arange(0, 10, 0.2)

# Notre fonction f(x) = 2*cos(x)+4
real_function = 2*np.cos(X_train) + 4

# J'ajoute du bruit pour construire notre y_train
y_train = real_function + np.random.normal(0,noise_level,len(X_train)) 

N = 100
# J'ai cr√©er 7 'hidden layer' avec 100 neuronnes dans chaque couche
regr = MLPRegressor(hidden_layer_sizes= tuple([N]*7)).fit(f(X_train), y_train)
predicted_func = regr.predict(f(X_train))
```

```python
plt.figure(figsize=(13, 3))
plt.plot(X_train, real_function, "-g", label="Real function")
plt.plot(X_train, predicted_func, "-r", label="Approximated function")
plt.legend(loc="best")
plt.show()
```
